```python
import datetime
import random
import string

# TEMPLATE DATA BUKU
buku_template = {
    'judul': 'judul buku',
    'penulis': 'nama penulis',
    'tahun': 0000,
    'tanggal_terakhir_pinjam': datetime.datetime(1111,1,11),
    'status': 'Tersedia'
}

# DATABASE PERPUSTAKAAN
data_buku = {}

while True:
    print(f"{' PERPUSTAKAAN ':^50}")
    print("-"*50)
    print("1. Tambah Buku")
    print("2. Tampilkan Daftar Buku")
    print("3. Keluar")
    print("-"*50)

    pilihan = input("Pilih menu (1/2/3): ")

    # MENU TAMBAH BUKU
    if pilihan == "1":
        buku = dict.fromkeys(buku_template.keys())

        buku['judul'] = input("Judul Buku        : ")
        buku['penulis'] = input("Penulis Buku      : ")
        buku['tahun'] = int(input("Tahun Terbit      : "))
        tahun = int(input("Masukkan tahun terakhir dipinjam (YYYY): "))
        bulan = int(input("Masukkan bulan terakhir dipinjam (1-12): "))
        tanggal = int(input("Masukkan tanggal terakhir dipinjam (1-31): "))
        buku["tanggal_terakhir_dipinjam"] = datetime.datetime(tahun, bulan, tanggal)
        buku['status'] = "Tersedia"

        # Generate KEY unik acak
        KEY = ''.join(random.choice(string.ascii_uppercase) for _ in range(6))
        data_buku.update({KEY: buku})

        print(f"\nBuku '{buku['judul']}' berhasil ditambahkan!\n")

    # MENU TAMPILKAN BUKU
    elif pilihan == "2":
        if len(data_buku) == 0:
            print("\nBelum ada buku yang terdaftar.\n")
        else:
            print("\nDAFTAR BUKU PERPUSTAKAAN")
            print("-"*70)
            print(f"{'KEY':<8} {'JUDUL':<20} {'PENULIS':<18} {'TAHUN':<6} {'STATUS':<12}")
            print("-"*70)

            for kode, buku in data_buku.items():
                print(f"{kode:<8} {buku['judul']:<20} {buku['penulis']:<18} {str(buku['tahun']):<6} {buku['status']:<12}")
            print("\n")

    # MENU KELUAR
    elif pilihan == "3":
        print("\nTerima kasih telah menggunakan sistem!")
        break

    else:
        print("\nPilihan tidak valid!\n")
```
