```python
Keranjang = {
    'apel':5000,
    'pisang':3000,
    'jeruk':7000,
    'anggur':15000,
}
print(Keranjang)

total = 0
for buah, harga in Keranjang.items():
  total += harga
  print(f'Harga {buah}: Rp{harga}')
print(f'Total harga belanja: {total}')

```
