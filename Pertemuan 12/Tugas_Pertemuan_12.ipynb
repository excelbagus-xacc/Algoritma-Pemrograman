```python
def luas_persegi_panjang():
  print('Menghitung Luas Persegi Panjang')
  p = int(input('Panjang\t= '))
  l = int(input('Lebar\t= '))
  luas = p * l
  return luas

def volume_balok():
  print('Menghitung Volume Balok')
  p = int(input('Panjang\t= '))
  l = int(input('Lebar\t= '))
  t = int(input('Tinggi\t= '))
  volume = p * l * t
  return volume

def luas_lingkaran():
  print('Menghitung Luas Lingkaran')
  r = int(input('Jari-jari\t= '))
  phi = 3.14
  luas = phi * r * r
  return luas

def volume_bola():
  print('Menghitung Volume Bola')
  r = int(input('Jari-jari\t= '))
  phi = 3.14
  bola = (4/3) * phi * r ** 3
  return bola

print(volume_balok())
```
